# Gradient Descent Regressor
1. Effect of learning rate
2. Effect of epochs

**Note:** The value of m is constant. 

```python
import numpy as np

from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
X, y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=25)
```

```python
plt.scatter(X, y)
```

```python
model = LinearRegression()
model.fit(X, y)
print(model.coef_)
print(model.intercept_)
```

```python
class GDRegressor:
    def __init__(self, learning_rate, epochs):
        self.m = 84.47
        self.b = -120
        self.lr = learning_rate
        self.epochs = epochs
    def fit(self, X, y):
        for i in range(self.epochs):
            slope = -2 * np.sum(y - self.m*X.ravel() - self.b)
            self.b = self.b - self.lr * slope
            print(slope, self.b)
        print(self.b)

```

```python
gd = GDRegressor(0.001, 100)
```

```python
gd.fit(X, y)
```
