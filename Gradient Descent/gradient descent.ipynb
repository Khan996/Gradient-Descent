```python
from sklearn.datasets import make_regression
import numpy
```

```python
X, y = make_regression(n_samples=4, n_features=1, noise=80, n_informative=1, n_targets=1, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X), color="red")
plt.show()
```

```python
# b = 0
y_pred = 78.35 * X + 0
y_pred
```

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X), color="red", label="LR")
plt.plot(X, y_pred, color="blue", label="b=0")
plt.legend()
```
